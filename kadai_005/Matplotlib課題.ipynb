```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 3, tight_layout=True)

y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)

x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1)

data = np.random.normal(15, 5, 2000)
plt.xlim(-10, 40)
axes[0, 2].hist(data)

x2 = 4 + np.random.normal(0, 2, 60)
y2 = 4 + np.random.normal(0, 2, len(x2))
axes[1, 0].scatter(x2, y2)

plt.show()
```
